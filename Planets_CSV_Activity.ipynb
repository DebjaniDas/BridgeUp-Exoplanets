```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy import units as u
from astropy.coordinates import SkyCoord
%matplotlib inline 
#percent sign gives command to jupyter notebook, not a python command: it will allow jupyter to show the plot
```

```python
df = pd.read_csv('planets.csv')
```

```python
df
```

```python
stars = df["st_mass"] <= 0.6
```

```python
df[stars]
```

```python
exodf = df.loc[stars]
```

```python
mass = exodf['st_mass']
```

```python
plt.hist(mass, bins = 301)
plt.title("Star Mass Histogram")
plt.xlabel("Mass")
plt.ylabel('Frequency')
```

```python
ra = exodf['ra']
```

```python
plt.hist(mass)
plt.hist(ra, alpha = 0.5) 
#alpha makes it see through!
plt.xlabel ('x')
plt.ylabel ('y')
#plots two histograms together!
```

```python
dec = exodf['dec']
```

```python
fig, ax = plt.subplots()
ax.scatter(ra, dec, c= "purple", marker = "*", alpha = 0.1, edgecolors = "green")
plt.title("RA and DEC Scatter Plot")
plt.xlabel("RA")
plt.ylabel("DEC")
```

```python
c = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')
```

```python
ra_rad = c.ra.radian 
#^ this converts the RA from our c variable (c.ra) into radians

dec_rad = c.dec.radian
#^ this converts the Dec from our c variable (c.dec) into radians
```

```python
pi = np.pi #python doesn't automatically know the number for pi, so we need to tell it
```

```python
ra_rad[ra_rad > pi] -= 2. * pi 
```

```python
fig = plt.figure(figsize=[10, 8]) 
# ^ This controls the size of the figure

ax = fig.add_subplot(111, facecolor='Black', projection = "aitoff")
# ^ This gives the background color

ax.scatter(ra_rad, dec_rad, marker='*')   # This plots all the points as stars
plt.grid(True)  # Turns on the grid in the back of the plot

#but these projections of the sky fail because they don't show you the distance the stars are from the earth
#there is a cluster of stars because those are the stars that Kepler recorded, and Kepler stares at one spot; the other stars are other stars that the NASA data contained
```
